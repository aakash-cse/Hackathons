```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.head(10)
```

```python
#  Our final submission file contains the following columns Item_Identifier	Outlet_Identifier Item_Outlet_Sales	
# we need to predict the Item_Outlet_Sales
test.head(5)
```

```python

train.drop('Item_Identifier',inplace=True,axis=1)
sub = test[['Item_Identifier','Outlet_Identifier']]
test.drop('Item_Identifier',inplace=True,axis=1)

```

```python
train['Item_Fat_Content'] = train['Item_Fat_Content'].map({'Low Fat':0, 'Regular':1, 'low fat':0, 'LF':0, 'reg':1})
test['Item_Fat_Content'] = test['Item_Fat_Content'].map({'Low Fat':0, 'Regular':1, 'low fat':0, 'LF':0, 'reg':1})

train['Outlet_Location_Type']=train['Outlet_Location_Type'].map({'Tier 1':1, 'Tier 3':3, 'Tier 2':2})
test['Outlet_Location_Type']=test['Outlet_Location_Type'].map({'Tier 1':1, 'Tier 3':3, 'Tier 2':2})

train['Outlet_Type']=train['Outlet_Type'].map({'Supermarket Type1':1, 'Supermarket Type2':2, 'Grocery Store':0, 'Supermarket Type3':3})
test['Outlet_Type']=test['Outlet_Type'].map({'Supermarket Type1':1, 'Supermarket Type2':2, 'Grocery Store':0, 'Supermarket Type3':3})

train['Outlet_Size']=train['Outlet_Size'].map({'Medium':2,'High':3, 'Small':1})
test['Outlet_Size']=test['Outlet_Size'].map({'Medium':2,'High':3, 'Small':1})

train['Outlet_Identifier']=train['Outlet_Identifier'].map({'OUT049':10, 'OUT018':5, 'OUT010':1, 'OUT013':2, 'OUT027':6, 'OUT045':8,'OUT017':3, 'OUT046':9, 'OUT035':7, 'OUT019':4})
test['Outlet_Identifier']=test['Outlet_Identifier'].map({'OUT049':10, 'OUT018':5, 'OUT010':1, 'OUT013':2, 'OUT027':6, 'OUT045':8, 'OUT017':3, 'OUT046':9, 'OUT035':7, 'OUT019':4})

train['Item_Type']=train['Item_Type'].map({'Dairy':0, 'Soft Drinks':1, 'Meat':15, 'Fruits and Vegetables':2,'Household':3,'Baking Goods':4,'Snack Foods':5, 'Frozen Foods':6, 'Breakfast':7, 'Health and Hygiene':8, 'Hard Drinks':9, 'Canned':10, 'Breads':11, 'Starchy Foods':12, 'Others':14, 'Seafood':13})
test['Item_Type']=test['Item_Type'].map({'Dairy':0, 'Soft Drinks':1, 'Meat':15, 'Fruits and Vegetables':2,'Household':3,'Baking Goods':4,'Snack Foods':5, 'Frozen Foods':6, 'Breakfast':7, 'Health and Hygiene':8, 'Hard Drinks':9, 'Canned':10, 'Breads':11, 'Starchy Foods':12, 'Others':14, 'Seafood':13})
```

```python
# now let us fill the null value Item_Type with others value
train['Item_Weight'].fillna(0,inplace=True)
test['Item_Weight'].fillna(0,inplace=True)

```

```python
#Similarly if outlet type is based on outlet size 
train['Outlet_Size'] = np.where(train.Outlet_Size.isnull()&train['Outlet_Type']==0,1,train['Outlet_Size'])
train['Outlet_Size'] = np.where(train.Outlet_Size.isnull()&train['Outlet_Type']==1,3,train['Outlet_Size'])
train['Outlet_Size'] = np.where(train.Outlet_Size.isnull()&train['Outlet_Type']==2,2,train['Outlet_Size'])
train['Outlet_Size'] = np.where(train.Outlet_Size.isnull()&train['Outlet_Type']==3,1,train['Outlet_Size'])


test['Outlet_Size'] = np.where(test.Outlet_Size.isnull()&test['Outlet_Type']==0,1,test['Outlet_Size'])
test['Outlet_Size'] = np.where(test.Outlet_Size.isnull()&test['Outlet_Type']==1,3,test['Outlet_Size'])
test['Outlet_Size'] = np.where(test.Outlet_Size.isnull()&test['Outlet_Type']==2,2,test['Outlet_Size'])
test['Outlet_Size'] = np.where(test.Outlet_Size.isnull()&test['Outlet_Type']==3,1,test['Outlet_Size'])
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(train.drop('Item_Outlet_Sales',axis=1),train.Item_Outlet_Sales,test_size=0.2)
```

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVR
from sklearn.linear_model import PassiveAggressiveRegressor
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import StackingRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.linear_model import LogisticRegression

est = LogisticRegression()
estimator = SVR(kernel="linear")
selector = RFE(est, n_features_to_select=5, step=1)
selector = selector.fit(x_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import ElasticNet
from sklearn.pipeline import make_pipeline
est = make_pipeline(StandardScaler(), ElasticNet())
est.fit(x_train)
est.predict(x_test)
```

```python
print(selector.score(x_test,y_test))
```

```python
sub['Item_Outlet_Sales']=HistGradientBoostingRegressor().fit(x_train,y_train).predict(test)
y_pred=HistGradientBoostingRegressor().fit(x_train,y_train).predict(x_test)
```

```python
sub.to_csv('sub.csv',index=False)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred))
```
